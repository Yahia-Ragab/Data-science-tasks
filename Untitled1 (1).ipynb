```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn as sk
sns.set()
```

```python
df =pd.read_csv('used_car_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df['AskPrice'] = df['AskPrice'].replace('₹ ', '', regex=True)
df['AskPrice'] = df['AskPrice'].replace(',', '', regex=True)
df['AskPrice'] = df['AskPrice'].astype(float)
print(df.head())
```

```python
print(df[df['kmDriven'].isna()])
```

```python
df = df.dropna(subset=['kmDriven'])
df.info()
```

```python
df['kmDriven'] = df['kmDriven'].replace(' km', '', regex=True)
df['kmDriven']=df['kmDriven'].replace(',','',regex=True)
df['kmDriven']=df['kmDriven'].astype(float)
df.head()
```

```python
df.info()
```

```python
print(df['Transmission'].unique(),
df['Owner'].unique(),
df['FuelType'].unique(),df['Brand'].unique(),len(df['Brand'].unique()))
```

```python
df['Transmission'] = df['Transmission'].map({'Manual': 0, 'Automatic': 1})
```

```python
df['Owner']=df['Owner'].map({'first':0,'second':1})
```

```python
newDf = df[['Age','kmDriven','Transmission','Owner','FuelType','AskPrice']]
newDf.head()
```

```python
newDf = pd.concat([newDf.drop('FuelType', axis=1), pd.get_dummies(newDf['FuelType'], dtype=int)], axis=1)
newDf.head()
```

```python
transmission_counts = newDf['Transmission'].value_counts()
transmission_labels = ['Manual', 'Automatic']
owner_counts = newDf['Owner'].value_counts()
owner_labels = ['First Owner', 'Second Owner']
print(transmission_counts, owner_counts)
```

```python
plt.figure(figsize=(6, 6))
plt.pie(transmission_counts, labels=transmission_labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Transmission Types')
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.pie(owner_counts, labels=owner_labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Owner Types')
plt.show()
```

```python
correlation = newDf[['Age','kmDriven','Transmission','Owner', 'Diesel','Hybrid/CNG','Petrol','AskPrice']].corr()
correlation_with_askprice = correlation['AskPrice']
print(abs(correlation_with_askprice))
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

x = newDf[['Age','kmDriven','Transmission','Owner', 'Diesel','Hybrid/CNG','Petrol']]
y = newDf['AskPrice']

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared: {r2}")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.5)
plt.title('Predicted vs Actual AskPrice')
plt.xlabel('Actual AskPrice')
plt.ylabel('Predicted AskPrice')
plt.show()
```
